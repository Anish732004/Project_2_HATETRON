```python
#@title Default title text
from graphviz import Digraph

dot = Digraph()

dot.node('W', 'Models used and their accuracy')
dot.node('L', 'Logistic Regression')
dot.node('R', 'Random Forest')
dot.node('M', 'Multinomial Naive Bayes')
dot.edges(['WL', 'WR', 'WM'])

dot.node('a', 'Accuracy = 90%')
dot.node('b', 'Accuracy = 90%')
dot.node('c', 'Accuracy = 82%')
dot.edges(['La', 'Rb', 'Mc'])

dot.render('output/model.jpg', view = False)
dot
```
